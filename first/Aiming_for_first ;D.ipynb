```python
import pandas as pd
import glob
from collections import Counter
```

```python
print(glob.glob("*"))
```

```python
train = pd.read_csv("train.csv")
print("Total No of Columns", len(train.columns))

nan_counter = Counter()
for column in  train.columns:
    nan_counter[column] = train[column].isnull().sum()
    
columns_with_no_nan = [x for x in nan_counter if nan_counter.get(x)==0]
print("len of columns with no null's",len(columns_with_no_nan))

no_nan_train = pd.DataFrame(train[columns_with_no_nan])
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
numeric_train = no_nan_train.select_dtypes(include=numerics)
numeric_columns = numeric_train.columns
categorical_columns = [x for x in columns_with_no_nan if x not in numeric_columns]
print("In that len of numeric columns", len(numeric_train.columns))
```
